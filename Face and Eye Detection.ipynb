###  "Facial Feature Detection with Haar Cascades"
### "Detecting Faces and Eyes in Images using OpenCV"

### Note: This is face detection NOT face recognition. We are only detecting if a face is in an image, not who the face actually is.

### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Read the image

```python
img=cv2.imread('kach.jpg',0)
img1=cv2.imread('kachh.jpg',0)
img2=cv2.imread('ju.jpg',0)
```

### Display the image

```python
plt.imshow(img, cmap = 'gray')
```

```python
plt.imshow(img1, cmap = 'gray')
```

```python
plt.imshow(img2, cmap = 'gray')
```

### Face Detection

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor = 1.2, minNeighbors = 12) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 2) 
        
    return face_img
```

```python
result = detect_face(img1)
```

```python
plt.imshow(result, cmap = 'gray')
```

```python
result = detect_face(img)
```

```python
plt.imshow(result, cmap = 'gray')
```

### Eye Detection

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    
    face_img = img.copy()
  
    eyes = eye_cascade.detectMultiScale(face_img, scaleFactor = 1.2, minNeighbors = 9) 
    
    
    for (x,y,w,h) in eyes: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 3) 
        
    return face_img
```

```python
result1 = detect_eyes(img2)
plt.imshow(result1, cmap = 'gray')
```
